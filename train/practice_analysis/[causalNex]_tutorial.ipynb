```python
import warnings
from causalnex.structure import StructureModel

warnings.filterwarnings("ignore")

sm = StructureModel()
```

```python
# 구조 시각화

from causalnex.plots import plot_structure, NODE_STYLE, EDGE_STYLE
import matplotlib.pyplot as plt
import networkx as nx
import os, webbrowser

viz = plot_structure(
    sm,
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)

viz.show('structure.html')
```

```python
# data definition

import pandas as pd

df = pd.read_csv('/home/yoo9519/pythonScript/datasets/student/student-por.csv', delimiter=';')
print(df.info())
df.head(10)
```

```python
drop_col = ['school','sex','age','Mjob', 'Fjob','reason','guardian']
data = df.drop(columns=drop_col)
data.head(5)
```

변수의 수가 증가하거나, 도메인 지식이 없을 경우, 구조를 수동으로 정의하는 것이 부정확(불가능) 하다.  
그래서 CausalNex를 사용하여, 데이터에서 구조 모델을 학습할 수 있다.  
본 문서에서 사용한 구조 학습 알고리즘은 NOTEARS 알고리즘임

구조를 사용할 때, 전체 데이터 세트를 사용할 수는 없지만, 기계 학습과 도메인 전문가 간의 공동 노력으로 간주되어야 하므로 항상 학습/테스트 분할을 사용할 필요는 없다.

```python
# 데이터 전처리: NOTEARS를 사용하기 위해, 숫자가 아닌, 변수를 인코딩하는 레이블로 변환해야 함

import numpy as np

struct_data = data.copy()
non_numeric_columns = list(struct_data.select_dtypes(exclude=[np.number]).columns)

print(non_numeric_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in non_numeric_columns:
    struct_data[col] = encoder.fit_transform(struct_data[col])
    
struct_data.head(5)
```

```python
# NOTEARS 알고리즘으로 학습
from causalnex.structure.notears import from_pandas

sm = from_pandas(struct_data)
```

```python
# plot함수로 structureModel을 시각화

viz = plot_structure(
    sm,
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)

viz.toggle_physics(False)
viz.show('fully_connected_structure.html')
```

```python
# threshold를 설정하여, 가장자리를 제거함

sm.remove_edges_below_threshold(0.8)
viz = plot_structure(
    sm,
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)

viz.show("threshold connected node.html")
```
